Show gender vs average scores visually with a simple, clean bar chart.

```python
import matplotlib.pyplot as plt

# Calculate averages
gender_avg = df.groupby('gender')[['math score','reading score','writing score']].mean()

# Plot
gender_avg.plot(kind='bar', figsize=(8,5))
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()
```

Female students outperform male students in reading and writing, suggesting stronger language skills on average, while males show slightly higher math performance.
